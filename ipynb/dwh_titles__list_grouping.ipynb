```python
############################################ 
### import festival lists
############################################ 

import os
import pandas as pd

data_path = r"G:/My Drive/casablanca/rpi/prod/interfaces/rpi10008_imdb__editor_lists/import/archive"
var_list = ("Const", "Title", "Description", "Directors", "Genres", "Release Date", "URL")

def load_festival_files(pattern, festival_name, festival_url, start_year, end_year):
    frames = []
    for year in range(start_year, end_year + 1):
        file_name = pattern.format(year=year)
        file_path = os.path.join(data_path, file_name)
        if os.path.exists(file_path):
            dfx = pd.read_csv(file_path, encoding="latin-1")
            dfx = dfx.loc[:, var_list]
            dfx["festival_name"] = festival_name
            dfx["festival_year"] = year
            dfx["festival_url"] = festival_url
            frames.append(dfx)
    result = pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()
    result.rename(columns={"Const": "tconst"}, inplace=True)
    return result

fst = load_festival_files(
    "Cannes Film Festival {year}.csv",
    "Cannes",
    "https://www.festival-cannes.com",
    2021,
    2029,
)

fst1 = load_festival_files(
    "Yesh{year}.csv",
    "Yesh!",
    "https://www.yesh.ch/",
    2021,
    2029,
)

fst2 = load_festival_files(
    "Zurich Film Festival {year}.csv",
    "ZFF",
    "https://zff.com/en/programme/",
    2021,
    2029,
)

fst3 = load_festival_files(
    "Sundance {year} Movie Guide.csv",
    "Sundance",
    "https://festival.sundance.org/",
    2020,
    2029,
)

fst4 = load_festival_files(
    "SXSW {year} Festival Guide.csv",
    "SXSW",
    "https://www.sxsw.com/festivals/film",
    2020,
    2029,
)

fst6 = load_festival_files(
    "Toronto International Film Festival {year}.csv",
    "TIFF",
    "https://www.tiff.net/",
    2020,
    2029,
)

fst8 = load_festival_files(
    "{year} Oscars - Academy Awards predictions.csv",
    "Oscars",
    "https://www.oscars.org/oscars/ceremonies",
    2021,
    2029,
)

fst9 = load_festival_files(
    "{year} Venice International Film Festival.csv",
    "Venice",
    "https://www.labiennale.org/en/cinema/",
    2021,
    2029,
)

fst10 = load_festival_files(
    "{year} Berlin International Film Festival.csv",
    "Berlin",
    "https://www.berlinale.de/en/home.html",
    2021,
    2029,
)

fst11 = load_festival_files(
    "Karolvy Vary {year}.csv",
    "KVIFF",
    "https://www.kviff.com",
    2021,
    2029,
)



fst3 = load_festival_files(
    "Sundance {year} Movie Guide.csv",
    "Sundance",
    "https://festival.sundance.org/",
    2020,
    2029,
)

fst4 = load_festival_files(
    "SXSW {year} Festival Guide.csv",
    "SXSW",
    "https://www.sxsw.com/festivals/film",
    2020,
    2029,
)

fst6 = load_festival_files(
    "Toronto International Film Festival {year}.csv",
    "TIFF",
    "https://www.tiff.net/",
    2020,
    2029,
)

fst8 = load_festival_files(
    "{year} Oscars - Academy Awards predictions.csv",
    "Oscars",
    "https://www.oscars.org/oscars/ceremonies",
    2021,
    2029,
)

fst9 = load_festival_files(
    "{year} Venice International Film Festival.csv",
    "Venice",
    "https://www.labiennale.org/en/cinema/",
    2021,
    2029,
)

fst10 = load_festival_files(
    "{year} Berlin International Film Festival.csv",
    "Berlin",
    "https://www.berlinale.de/en/home.html",
    2021,
    2029,
)

fst11 = load_festival_files(
    "Karolvy Vary {year}.csv",
    "KVIFF",
    "https://www.kviff.com",
    2021,
    2029,
)
fst0 = load_festival_files(
    "{year} - The year in movies.csv",
    "YTD",
    "https://www.kulturinfo.ch/kino/index.html",
    2021,
    2029,
)

 
print(fst.shape, fst0.shape)
fst0.tail(3)
```

```python
############################################ 
### import current screening list
############################################ 

from datetime import datetime

def load_current_screening_lists(filename, festival_name, festival_year, festival_url):
    file_path = os.path.join(data_path, filename)
    if not os.path.exists(file_path):
        return pd.DataFrame()
    df = pd.read_csv(file_path, encoding="latin-1")
    df = df.loc[:, var_list]
    df["festival_name"] = festival_name
    df["festival_year"] = festival_year
    df["festival_url"] = festival_url
    df.rename(columns={"Const": "tconst"}, inplace=True)
    return df


current_year = datetime.now().year

fstw = load_current_screening_lists(
    "whatson.csv",
    "WhatsOn",
    current_year,
    "https://www.kulturinfo.ch/kino/index.html",
)

print(fstw.shape)
fstw.tail(3)
```
